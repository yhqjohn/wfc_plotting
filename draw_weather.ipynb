```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline


cm = plt.cm.get_cmap('tab10')
markers = ['o', '^', 's', 'X', 'D']
```

```python
df = pd.read_csv('la_salada.csv')
ax = df.plot(x="Hour", y="GHI", legend=False, c=cm(0), marker=markers[0])
plt.ylabel('GHI [$W/m^2$]')
plt.xlabel('Time')
ax2 = ax.twinx()
df.plot(x="Hour", y="Temperature", ax=ax2, legend=False, c=cm(1), marker=markers[1])
plt.ylabel('Temperature [â„ƒ]')
ax.figure.legend()
plt.savefig('weather.svg')
plt.show()
```
